```python
import datetime
class Person :
    def __init__(self,name,date) :
        self.name = name
        self.date = datetime.datetime.strptime(date, "%d-%m-%Y")
    
    def calculate_age(self):
        import datetime
        current_date = datetime.datetime.now()
        age = current_date.year - self.date.year
        return age
    
    def get_weekday_name(self):
        return self.date.strftime("%A")
    
    
def main():
    people = []
    total_people = 0
    sunday_birthdays = []

    while True:
        input_str = input("Enter the name accompanied by date of birth(dd-mm-yyyy) : ")
        if not input_str:
            break
        try:
            name, birthdate_str = input_str.split(", ")
            person = Person(name, birthdate_str)
            age = person.calculate_age()
            weekday_name = person.get_weekday_name()
            
            people.append(person)
            total_people += 1
            
            if weekday_name == "Sunday":
                sunday_birthdays.append(name)
            
            print(f"{name} is {age} years old and she/he was born on {weekday_name}")
        
            
        except ValueError:
            print("Invalid date: ", input_str)
            
    if people:
        oldest_person = max(people, key=lambda person: person.calculate_age())
        youngest_person = min(people, key=lambda person: person.calculate_age())
        print("\nThe oldest one is", oldest_person.name)
        print("The youngest one is", youngest_person.name)
    else:
        print("\nThere is no oldest or youngest person.")
    
    print("Total People:", total_people)
    
    # Print people born on Sunday
    if sunday_birthdays:
        print("\nPeople born on Sunday:", ", ".join(sunday_birthdays))
    else:
        print("\nNo one was born on Sunday.")

if __name__ == "__main__":
    main()
```
